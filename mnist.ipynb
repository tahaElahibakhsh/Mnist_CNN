## <div class="custom-section"><div class="custom-header">Step 1.2 | data loading</div><div class="custom-content">at first we import the mnist shits from keras.datasets and visualize it</div></div>


```python
# tensors has vectors in between 0-255 and we need to take it in 0-1 so:
(x_train, y_train), (x_test, y_test) = keras.datasets.mnist.load_data()
print(x_train[0]) # to show example
x_train = x_train.astype('float32') / 255.
x_test = x_test.astype('float32') / 255.
x_train = x_train.reshape((*x_train.shape, 1))
x_test = x_test.reshape((*x_test.shape, 1))
y_train = tf.keras.utils.to_categorical(y_train)
y_test = tf.keras.utils.to_categorical(y_test)
print(x_train[0]) # to show example after
```

```python
print(f"X_train.shape: {x_train.shape}")
print(f"y_train.shape: {y_train.shape}")
print(f"X_test.shape: {x_test.shape}")
print(f"y_test.shape: {y_test.shape}")
```

```python
plt.imshow(x_train[3].reshape(28, 28), cmap='gray')
plt.title(f"label: {np.argmax(y_train[3])}")
plt.show()
```

```python
# visualize some of them
for i in range(9):
    plt.subplot(331 + i)
    plt.imshow(x_train[i].reshape(28, 28), cmap=plt.get_cmap('gray'))
plt.show()
```

## <div class="custom-section"><div class="custom-header">Step 2.1 |  model structure</div><div class="custom-content">design a cnn model with one 16,4,4 conv layer and one pooling layer and for th main tune one fully connected layer</div></div>


```python
model = keras.Sequential()
model.add(keras.layers.Conv2D(16, (4,4), activation="relu", input_shape=(28,28,1)))
model.add(keras.layers.MaxPool2D((2,2)))
model.add(keras.layers.Dropout(0.25))

model.add(keras.layers.Flatten())
model.add(keras.layers.Dense(32, activation="relu"))
model.add(keras.layers.Dropout(0.5))

model.add(keras.layers.Dense(10, activation='softmax'))
model.summary()

opt = tf.keras.optimizers.SGD(learning_rate=0.01, momentum=0.9)
model.compile(optimizer=opt, loss='categorical_crossentropy', metrics=['accuracy', 'f1_score'])
```

## <div class="custom-section"><div class="custom-header">Step 2.3 | model evaluation</div><div class="custom-content">confusion matrix and classification report of model</div></div>


```python
import seaborn as sns

en_model = keras.models.load_model("en_model.keras")

y_pre = en_model.predict(x_test)
y_pre = np.argmax(y_pre, axis = 1)
y_test_ne = np.argmax(y_test, axis = 1)
c_matrix = (confusion_matrix(y_pre, y_test_ne))
plt.figure(figsize=(10, 10))
sns.heatmap(c_matrix, annot=True, cmap="coolwarm")
plt.ylabel("real label")
plt.xlabel("predicted label")
print(classification_report(y_pre, y_test_ne))
```

```python
wrong_idx = [i for i, (yp, yt) in enumerate(zip(y_pre, y_test_ne)) if yp != yt]

print(f"{len(wrong_idx)} samples misclassified")

plt.figure(figsize=(10, 6))
for idx, wrong in enumerate(wrong_idx[:20]):
    plt.subplot(5, 4, idx + 1)
    plt.imshow(x_test[wrong].reshape(28, 28), cmap='gray')
    plt.title(f"True: {y_test_ne[wrong]}\nPred: {y_pre[wrong]}")
    plt.axis('off')

plt.tight_layout()
plt.show()
```

## <div class="custom-section"><div class="custom-header">Step 2.3 |  model test</div><div class="custom-content">test the model with handwreited number</div></div>


```python
from PIL import Image
import matplotlib.pyplot as plt

def load_image_for_model(image_path):
    img = Image.open(image_path).convert("L")
    img = img.resize((28, 28), Image.Resampling.LANCZOS)
    arr = np.array(img).astype("float32") / 255.0
    arr = arr.reshape(1, 28, 28, 1)   # (batch, h, w, c)
    return arr, img

x, img = load_image_for_model("four.png")

en_model = keras.models.load_model("en_model.keras")
pred = en_model.predict(x)
print("neural net prediction:", np.argmax(pred))

plt.imshow(img, cmap="gray")
plt.title("generated img")
plt.axis("off")
plt.show()
```
